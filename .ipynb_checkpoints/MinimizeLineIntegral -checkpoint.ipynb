```python
def flatten(lol):
    fl = []
    for l in lol:
        fl.extend(l)
    return fl

def set_seed(seed):
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)

set_seed(1)
```

```python
at_model = common.state.State.load('examples/readme/cifar10_at/classifier.pth.tar')
ccat_model = common.state.State.load('examples/readme/cifar10_ccat/classifier.pth.tar')
# msd_model = common.state.State.load('examples/readme/cifar10_msd/classifier.pth.tar')
normal_model = common.state.State.load('examples/readme/cifar10_normal/classifier.pth.tar')
models = [ccat_model.model, at_model.model, normal_model.model]
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def load_model(model):
    model.to(device)
    model.eval()
    for p in model.parameters():
        p.requires_grad = False
for model in models:
    load_model(model)
```

```python
batch_size = 64
train_dataset = torch.utils.data.DataLoader(common.datasets.Cifar10TrainSet(), batch_size=batch_size, shuffle=False, num_workers=0)
test_dataset = torch.utils.data.DataLoader(common.datasets.Cifar10TestSet(), batch_size=batch_size, shuffle=False, num_workers=0)
raw_test_dataset = common.datasets.Cifar10TestSet()
classes = ['airplane',
'automobile',
'bird',
'cat',
'deer',
'dog',
'frog',
'horse',
'ship',
'truck',]
```

```python
def get_pivots(source_image,target_image,num_pivots):
    pivots = [ ((num_pivots+1-n)*source_image +n*target_image)/(num_pivots+1) for n in range(1, num_pivots+1)]
    return pivots

def path_between_pivots(start,end,t,aparams,bparams,mask):
    im = torch.zeros(*start.shape).to(start.device)
    tvec = torch.Tensor([t**i for i in range(1, len(aparams[0])+1)]).to(start.device)
    for i, (ap, bp) in enumerate(zip(aparams, bparams)):
        mi = (mask == i)
        im +=  (start*mi) * torch.sum(ap*(1-tvec))/torch.sum(ap) + (end*mi) * torch.sum(bp*tvec)/torch.sum(bp)
    return im

def get_path(model,start,end, device,
                        num_pivots = 14,
                        num_iters = 1000,
                        num_points = 10,
                        beta = 10,
                        lr = 0.01,
                        conf_thresh = 0.1,
                        im_thresh = 0.3,
                        num_parts = 10,
                        degree=4):
    liveloss = PlotLosses()
    start = start.permute(2,0,1).to(device)
    end = end.permute(2,0,1).to(device)
    pivots = get_pivots(start, end, num_pivots)
    pivot_params = [torch.nn.Parameter(p).to(device) for p in pivots]
    aparams = [torch.nn.Parameter(torch.ones(num_parts, degree, device=device)) for pi in range(num_pivots+1)]
    bparams = [torch.nn.Parameter(torch.ones(num_parts, degree, device=device)) for pi in range(num_pivots+1)]
    mask = torch.randint(num_parts, start.shape, device=device)
    path_params = aparams+bparams
    opt = torch.optim.Adam(path_params+pivot_params, lr=lr)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(opt, mode='min', patience=20, verbose=True)
    print(f"Difference between source and target: {torch.norm(end - start)}")
    for i in range(num_iters):
#         full_img_path = torch.stack(
#                                 flatten([[path_between_pivots(s,e,t,app, bpp,mask) for t in np.sort(np.random.rand(num_points))] 
#                                  for s, e, app, bpp in zip([start] + pivot_params, pivot_params + [end], aparams, bparams)])
#                         )
        full_img_path = torch.stack(
                                flatten([[path_between_pivots(s,e,t,app, bpp,mask) for t in np.arange(num_points)/num_points] 
                                 for s, e, app, bpp in zip([start] + pivot_params, pivot_params + [end], aparams, bparams)])
                        )
        logits = model((full_img_path).clamp(0,1))
        pred_probs = softmax(logits)
        source_class = pred_probs[0].argmax()
        conf_diffs = pred_probs[0,source_class] - pred_probs[:,source_class]
        full_img_path = torch.cat((full_img_path, end.unsqueeze(0)), dim=0)
        img_diffs = torch.stack([torch.norm(x) for x in full_img_path[:-1] - full_img_path[1:]])
        loss = torch.norm( relu(conf_diffs- conf_thresh))**2 + (beta)*torch.norm(relu(img_diffs -im_thresh))**2
        scheduler.step(loss)
        opt.zero_grad()
        loss.backward()
        opt.step()
        logs = {'loss':loss, 'conf diff':conf_diffs.mean(), 'img diff':img_diffs.mean()}
        liveloss.update(logs)
        liveloss.send()
#         print(f"Loss at iteration {i} is {loss}")
#         print(f"Max conf diff at iteration {i} is {conf_diffs.max()}")
#         print(f"Max consecutive img diff at iteration {i} is {img_diffs.max()}")
    pivot_params = [p.detach() for p in pivot_params]
    path_params = [p.detach() for p in path_params]
   
    return path_params, pivot_params, conf_diffs, img_diffs, full_img_path
```

```python
def get_targets(models, start_ind=0, end_ind=10, indices=None, filt=lambda pred_label, i, max_prob: pred_label == i): 
    target_img_indices = []
    target_images = []
    if indices:
        target_img_indices = indices
    for i in range(start_ind, end_ind):
        if indices:
            target_images.append(dataset_per_class[i][indices[i]])
            continue
        while True:
            ind = random.randint(0, len(dataset_per_class[i]))
            source_image = common.torch.as_variable(dataset_per_class[i][[ind]], True)
            source_image = source_image.permute(0, 3, 1, 2).to(device)
            probs = [torch.nn.functional.softmax(model((source_image).clamp(0,1))) for model in models]
            pred_labels = [int(torch.argmax(prob, dim=-1)[0]) for prob in probs]
            if len(set(pred_labels)) == 1 and pred_labels[0] == i:
#             if filt(pred_labels, i, max_prob):
                target_img_indices.append(ind)
                target_images.append(dataset_per_class[i][ind])
                break
    return target_img_indices, target_images
```
